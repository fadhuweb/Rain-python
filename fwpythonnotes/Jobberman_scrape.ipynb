```python
with open('Jobs.txt', 'w', encoding ='utf-8') as f:
    for index in range(1,170):
        page = requests.get(f'https://jobberman.com/jobs/?page={index}')
        soup = BeautifulSoup(page.text, 'html.parser')
        divs = soup.find_all('body', class_ = 'customer jm-nigeria min-h-full') 

        for div in divs:
            job_name = div.find('p',  class_ ="text-lg font-medium break-words text-link-500").text
            company = div.find('a',  class_="text-loading-animate text-loading-animate-link").text
            job_conditions = div.find('span', class_ ="mb-3 px-3 py-1 rounded bg-brand-secondary-100 mr-2 text-loading-hide").text
            job_function = div.find('p', class_="text-sm text-gray-500 text-loading-animate inline-block").text
            price= div.find('span', class_='mr-1').text
            f.write(f'Job: {job_name}\nCompany: {company}\nConditions: {job_conditions}\nFunction:{job_function}\nPrice:{price}')
            f.write('\n')
        
```
