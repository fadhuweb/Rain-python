```python
import pandas as pd
import numpy as np
import os
```

```python
folder_path  = r"D:\AIML_PROGRAM\fwaimlnotes\Sales_Data\Sales_Data"
```

```python
all_df = pd.DataFrame()
for filename in os.listdir(folder_path):
    month_df = pd.read_csv(rf"{folder_path}\{filename}")
    all_df = pd.concat([all_df,month_df],axis=0)
all_df
```

```python
all_df['Quantity Ordered'] = pd.to_numeric(all_df['Quantity Ordered'],errors='coerce')
all_df['Price Each'] = pd.to_numeric(all_df['Price Each'],errors='coerce')
```
